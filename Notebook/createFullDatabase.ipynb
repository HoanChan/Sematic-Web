````python
lsPro = list(onto.properties())
def get_properties(obj):
    return [p for p in lsPro if obj.is_a[0] in p.domain] + [p for p in lsPro if obj in p.domain]
print('STT\tTên lớp\tKế thừa từ\tThuộc tính')
lsClass= list(onto.classes())
for i, class_ in zip(range(len(lsClass)), lsClass):
    print(f"{str(i+1)}\t{class_.name}\t{str(class_.is_a).split('.')[1][:-1]}\t{', '.join([x.name for x in get_properties(class_)])}")
    # print(str(i+1)+'.', class_.name, 'kế thừa từ', class_.is_a)
    # for x in get_properties(class_) :
    #     print('-', x.name, ':', x.range[0])
````

````python
print('\nSTT\tTên thuộc tính\tDomain\tRange\tLoại')
import re
i=1
for p in lsPro:
    if p.is_a[0].name == 'ObjectProperty': continue
    domains = set([str(d).split('.')[1] for d in p.domain])
    ranges = []
    for r in p.range:
        data = re.findall(r"[\w']+", str(r))
        ranges.append(data[1].replace("'", ''))
    stype = [a.name for a in p.is_a][1:]
    print(f"{i}\t{p.name}\t{', '.join(domains)}\t{', '.join(set(ranges))}\t{', '.join(stype)}")
    i+=1
````

````python
# Đọc CSDL từ file
onto = get_ontology("../app/data/education.owl").load()

# In DS học sinh ra màn hình
print("Danh sách học sinh:")
stt = 1
for hs in onto.HocSinh.instances():
    print(stt, hs.hoTen)
    stt += 1
````

````python
# In DS giáo viên
print("Danh sách giáo viên:")
for gv in onto.GiaoVien.instances():
    print(gv.hoTen, gv.ngaySinh, gv.gioiTinh, gv.trinhDo, getattr(gv.lopChuNhiem,'ten',''), *[mon.ten for mon in gv.dayMon])
````

````python
# In DS HS trong lớp 10C01
print("Danh sách học sinh lớp 10C01:")
for hs in onto.HocSinh.instances():
    if hs.hocLop.ten == "10C01":
        print(hs.hoTen, hs.ngaySinh, hs.gioiTinh, hs.hocLop.ten)
````

````python
# in ds GV chủ nhiệm của từng lớp
print("Danh sách giáo viên chủ nhiệm:")
for lop in onto.LopHoc.instances():
    gv = lop.giaoVienChuNhiem
    if gv:
        print(gv.hoTen, gv.ngaySinh, gv.gioiTinh, gv.trinhDo, getattr(gv.lopChuNhiem,'ten',''))
````

````python
# Sửa tên của giáo viên chủ nhiệm Nguyễn Văn X thành Nguyễn Văn Tèo
print("Sửa tên của giáo viên chủ nhiệm Nguyễn Văn X thành Nguyễn Văn Tèo")
for gv in onto.GiaoVien.instances():
    if gv.hoTen == "Nguyễn Văn X":
        gv.hoTen = "Nguyễn Văn Tèo"
        print(gv.hoTen, gv.ngaySinh, gv.gioiTinh, gv.trinhDo, getattr(gv.lopChuNhiem,'ten',''))
````

````python
print('- Liệt kê ban giám hiệu')
BGH = onto.search_one(type = onto.ToChuc, ten = 'Ban giám hiệu')
for nv in BGH.thanhVien:
    print(nv.hoTen, nv.chucVu[0].ten, nv.chucVu[0].phuCap)
print('- Liệt kê tổ Văn phòng')
VP = onto.search_one(type = onto.ToChuc, ten = 'Văn phòng')
for nv in VP.thanhVien:
    print(nv.hoTen, nv.chucVu[0].ten, nv.chucVu[0].phuCap)
````

````python
print('Liệt kê các tổ trưởng và tổ phó của các tổ chuyên môn')
toChuyenMon = [x for x in onto.search(type = onto.ToChuc) if x.name.startswith('To')]
for tc in toChuyenMon:
    print(tc.ten)
    for nv in tc.thanhVien:
        print('   -', nv.hoTen, nv.chucVu[0].ten, nv.chucVu[0].phuCap, nv.dayMon)
````

````python
print('Liệt kê các phòng kèm tổ chức sử dụng')
for p in onto.Phong.instances():
    print(p.ten, end = ' ')
    print('-', p.suDungBoi.ten if p.suDungBoi else '', end = ' ')
    print()
````

````python
print('Liệt kê các hoạt động')
for hd in onto.HoatDong.instances():
    print(hd.ten, hd.ngayBatDau, hd.ngayKetThuc,[gv.hoTen for gv in hd.quanLy], [ng.hoTen for ng in hd.thamGia])
````

````python
print('Liệt kê các HS điểm danh ở 1 hoạt động')
hd = onto.search(type = onto.HoatDong)[4]
dd = onto.search(type = onto.DiemDanh, hoatDong = hd)
for d in dd:
    print(d.nguoi.hoTen, d.ngayGio, d.trangThai)
````

````python
print('Liệt kê danh mục phòng kèm theo danh sách thiết bị trong phòng')
for p in onto.Phong.instances():
    print(p.ten, end=' ')
    dsTB = [tb for tb in onto.ThietBi.instances() if tb.phong == p]
    for tb in dsTB:
        print('\t-', tb.ten, tb.soLuong, end=' ')
    print()
````
